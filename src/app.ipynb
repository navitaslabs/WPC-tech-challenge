```python
import pandas
patient = pandas.read_csv('datasets/patients.csv',delimiter=';',encoding='utf-8')
conditions = pandas.read_csv('datasets/conditions.csv',delimiter=';',encoding='utf-8')
encounters = pandas.read_csv('datasets/encounters.csv',delimiter=';',encoding='utf-8')
medications = pandas.read_csv('datasets/medications.csv',delimiter=';',encoding='utf-8')
observations = pandas.read_csv('datasets/observations.csv',delimiter=';',encoding='utf-8')
len(patient.index)
```

```python
patient_unique = patient.drop_duplicates()
print('patients original rows:', len(patient),'.unique rows: ',len(patient_unique))
```

```python
patient_unique_combination = patient.drop_duplicates(subset=['Social_Sec_Num','First_Name','Surname','BIRTH_DATE'])
len(patient_unique_combination)
```

```python
conditions_unique = conditions.drop_duplicates()
print('conditions original rows:', len(conditions),'.unique rows: ',len(conditions_unique))
```

```python
encounters_unique = encounters.drop_duplicates()
print('encounters original rows:', len(encounters),'.unique rows: ',len(encounters_unique))
```

```python
medications_unique = medications.drop_duplicates()
print('medications original rows:', len(medications),'.unique rows: ',len(medications_unique))
```

```python
observations_unique = observations.drop_duplicates()
print('observations original rows:', len(observations),'.unique rows: ',len(observations_unique))
```
